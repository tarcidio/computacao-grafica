````python
# Importa bibliotecas
import glfw_functions
import glsl_functions
import load_models
import camera

# Cria janela glfw
window, width, heigh = glfw_functions.init_window("Trabalho 02")
# Cria programa glsl
program = glsl_functions.create_program_glsl()

# Ativando texturas 2D
glsl_functions.on_texture()
# Ativando 3D
glsl_functions.on_3d()
# Libera 10 id para texturas
glsl_functions.define_num_textures(10)

vertices, textures, normals, sun, objs_wave, objs_geometric, sky_radius = load_models.load_models_obj()

# Solicita espaço e armazena os vértices na GPU
glsl_functions.create_data_space(vertices, textures, normals, program)

# Criar camera
# Posicao inicial da camera
cam = camera.create_camera()

# Cria variável que define polygonal
polygonal = [True]

# Cores ambiente
color_amb = [1.0, 1.0, 1.0]

# Define eventos de teclado e mouse 
glfw_functions.setup_keyboard_event_handlers(window, color_amb, polygonal, cam, sun, objs_wave, objs_geometric, sky_radius, width, heigh)

# Exibe janela
glfw_functions.show_window(window)

# Loop principal
while not glfw_functions.verify_window(window):
    # Inicializa loop
    glfw_functions.init_main_loop_glfw(window)
    glsl_functions.init_main_loop_glsl()
    
    # Verifica se deve ser exibido as malhas
    glsl_functions.check_polygonal(polygonal)

    # Desenha objetos
    load_models.draw_objs(program, objs_geometric, objs_wave)

    # Atualiza cor ambiente
    glsl_functions.update_color(program, color_amb)

    # Envia dados da camera
    camera.send_data(program, cam, width, heigh)

# Finaliza sistema glfw
glfw_functions.finish_glfw()
````
